```python
rawvisadata=pd.read_csv("GMP_GlobalVisaCostDataset_v1.0.csv")
```

```python
#Look under the hood.
```

```python
rawvisadata.info()
```

```python
#I'd like less columns. Removing columns 2 and 3, which provide 'ISO' codes for each country. Also removing Column 19 and 21, 'Subregions'.
rawvisadata = rawvisadata.drop(['source_iso3','target_iso3','target_subregion', 'source_subregion'], axis=1)
```

```python
#Great, slightly less data.
```

```python
#Question One:
#What do the prices look like for visas for the United States?
```

```python
USentryvisadata = rawvisadata.loc[rawvisadata['target'] == 'United States of America']
```

```python
USentryvisadata.head()
```

```python
#sort to see which tourist visa to the US is the cheapest.
UStouristvisasorted = USentryvisadata.sort_values(by='tourist_visa', ascending=True)
```

```python
UStouristvisasorted.head()
```

```python
#Insight 1: Its cheaper to visit the United States from a rich state rather than a poor one.

#sort to see which student visa to the US is the cheapest
USstudentvisasorted = USentryvisadata.sort_values(by='student_visa', ascending=False)
USstudentvisasorted.head()
```

```python
#looks like they all cost $160, what about for a different country?
Franceentryvisadata = rawvisadata.loc[rawvisadata['target'] == 'France']
Francestudentvisa = Franceentryvisadata.sort_values(by='student_visa', ascending=True)
Francestudentvisa.head(30)
```

```python
#looks like student visas to France are free.
```

```python
#Lets try one more, Chile.
chiledata = rawvisadata.loc[rawvisadata['target'] == 'Chile']
chilestudentdata = chiledata.sort_values(by='student_visa', ascending=True)
chilestudentdata.head(30)
```

```python
#most expensive?
chilestudentdata = chiledata.sort_values(by='student_visa', ascending=False)
chilestudentdata.head(20)
```
